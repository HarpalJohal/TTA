```python
#-------------------------------------------------------------------------------
# Name:         HLT Week 3
# Author:       harpal johal
# Created:      24/07/2021
#
# Task 2 :      Write a program to ask a student for their percentage mark and converts this to a grade.
#               The conversion will be done in a function called mark_grade
# extension to Task 2
#               Ask the user for their target grade and print this with their mark
#               If their target grade > exam grade display a suitable message
#               If their target grade = exam grade display a suitable message
#               If their target grade < exam grade display a suitable message
# -------------------------------------------------------------------------------
def mark_grade(percentage_mark, target_grade):
    if percentage_mark <= 40:
        grade = "E"
    elif percentage_mark <= 55:
        grade = "D"
    elif percentage_mark <= 70:
        grade = "C"

    elif percentage_mark <= 85:
        grade = "B"
    elif percentage_mark <= 100:
        grade = "A"

    if target_grade < grade:
        grade_message = "Try harder, maybe next time!"
    elif target_grade == grade:
        grade_message = "Congratualtions on meeting your expectation!"
    else:
        grade_message = "Excellent, You achieved better than expected !"

    print(percentage_mark, "% is a Grade", grade,  grade_message)
    
    return grade

    
    

percentage_mark = int(input("What percentage was your pass mark? "))
#Extn to task 2
target_grade = (input("What was your target grade (A-E) "))

# convert_grade
mark_grade(percentage_mark,target_grade)
```
