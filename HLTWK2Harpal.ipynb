```python
#-------------------------------------------------------------------------------
# Name:         HLT Week 2
# Author:       harpal johal
# Created:      18/07/2021
#
# Task 5 :      Write a program which a user can select a tea or coffee. Then 
#               choose if they wish to add sugar or milk.  
#               Display message of drink ordered based on their choices.
#
# -------------------------------------------------------------------------------
#
#Select Drink Option
#
def OrderDrink(drink_opt):
    valid_opt = True

    while valid_opt:

        print ("""
        1. Tea
        2. Coffee

  
        """)

#Select User Request

        UserDrink = input("Which drink you would like: ")

#Check Selection
        if UserDrink == "1":
            drink_opt = " tea" 
            valid_opt = False
        elif UserDrink == "2":
            drink_opt = " coffee" 
            valid_opt = False
        else:
            print ("Invalid Drink Option.")

    return drink_opt

# Add Sugar?
#
def WantSugar(sugar_opt):

    valid_opt = True

    while valid_opt:
        AddSugar = input("Would you like to add sugar (Y/N): ")

        if AddSugar == "Y" or AddSugar == "y": 
            sugar_opt = " with sugar"
            valid_opt = False
        elif AddSugar == "N" or AddSugar == 'n':
            sugar_opt = " without sugar" 
            valid_opt = False
        else:
            print ("Invalid Sugar Option.")

    return sugar_opt

# Add Milk ?  
# # 
def WantMilk(milk_opt):
    valid_opt = True

    while valid_opt:
    
        AddMilk = input("Would you like to add milk (Y/N): ")

        if AddMilk == "Y" or AddMilk == "y": 
            milk_opt = " white"         
            valid_opt = False
        elif AddMilk == "N" or AddMilk == 'n':
            milk_opt = " black" 
            valid_opt = False
        else:
            print ("Invalid Milk Option.")

    return milk_opt
   
# Display drink order details
def PrintOrder(milk_opt, drink_opt,sugar_opt):
    milk = milk_opt
    drink = drink_opt
    sugar = sugar_opt

    Order = "You have ordered a" + milk + drink + sugar +". Enjoy your drink!"

    print(Order)

drink_opt = " "
sugar_opt = " "
milk_opt = " "

drink_opt = OrderDrink(drink_opt)
sugar_opt = WantSugar(sugar_opt)
milk_opt = WantMilk(milk_opt)

PrintOrder(milk_opt,drink_opt,sugar_opt)

```

```python
#-------------------------------------------------------------------------------
# Name:         HLT Week 2
# Author:       harpal johal
# Created:      19/07/2021
#
# Task 2 :      A motorbike costs £2000 and loses 10% of its value every year. Set up a function that performs the calculation 
#               by passing in parameters. Then using a loop, print the value of the bike every following year until it falls 
#               below £1000. 
# 
# -------------------------------------------------------------------------------
#
# function that takes in bike value and loss, to print value of bike each year until falls below £1000

MotorbikeCost = int(input("How much did your motorbike cost?"))
InpBikeValue = MotorbikeCost
InpAnnualLoss = int(input("what % is its annual loss?"))
Year = 1

# define function to calculate loss of motorbike
def calc_loss(InpBikeValue,InpAnnualLoss,Year):
    BikeValue = InpBikeValue
    AnnualLoss = InpAnnualLoss
    
    while BikeValue > 1000:
        print("The value of your motorbike is £", BikeValue, "in Year", Year)
        BikeValue = BikeValue - (BikeValue * (AnnualLoss/100))
        Year += 1

# perform loss calculation
calc_loss(InpBikeValue,InpAnnualLoss,Year)

```

```python
#-------------------------------------------------------------------------------
# Name:         HLT Week 2
# Author:       harpal johal
# Created:      19/07/2021
#
# Task 3 :      Write a program which will ask for two numbers from a user. Then 
#               offer a menu to the user giving them a choice of operator:
#               e.g. – Enter “a” if you want to add
#               “b” if you want to subtract
#               Include +, -, /, *, ** square (to the power of). Once the user has 
#               selected which operator they wish to use, perform the calculation 
# 
# -------------------------------------------------------------------------------
# This function adds two numbers
def add(x, y):
    return x + y
# This function subtracts two numbers
def subtract(x, y):
    return x - y
# This function muliplies two numbers
def multiply(x, y):
    return x * y
# This function divides two numbers
def divide(x, y):
    return x / y
# This function calculates x to the power of y
def topowerof(x, y):
    return x ** y

#Select Numbers
NumberOne = input("What is your first number?")
NumberOne = int(NumberOne)
NumberTwo = input("What is your second number?")
NumberTwo = int(NumberTwo)

UserChoice = True

while UserChoice:

# Display Menu
  print ("""
  Enter A to Add
  Enter B to Subtract
  Enter C to Divide
  Enter D to Multiply
  Enter E to Square

  Enter X to Exit
  """)

#Select User Request

  UserChoice = input("Please select option for the mathematical operation you wish to perform: ")

  if UserChoice in ('A','B', 'C', 'D', 'E', 'X', 'a', 'b', 'c', 'd', 'e', 'x'):
      if UserChoice == 'A' or UserChoice == 'a':
          print(NumberOne, "+", NumberTwo, "=", add(NumberOne,NumberTwo))
      elif UserChoice == 'B' or UserChoice == 'b':
          print(NumberOne, "-", NumberTwo, "=", subtract(NumberOne,NumberTwo))
      elif UserChoice == 'C' or UserChoice == 'c':
          print(NumberOne, "/", NumberTwo, "=", divide(NumberOne,NumberTwo))
      elif UserChoice == 'D' or UserChoice == 'd':
          print(NumberOne, "*", NumberTwo, "=", multiply(NumberOne,NumberTwo))
      elif UserChoice == 'E' or UserChoice == 'e':
          print(NumberOne, "to the power of", NumberTwo, "=", topowerof(NumberOne,NumberTwo))
      elif UserChoice == 'X' or UserChoice == 'x':
          break
      break  
  else:
      print ("Invalid Option Selected!")

```
